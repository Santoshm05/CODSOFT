```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
```

DATA PREPROCESSING

```python
spam.isnull().sum()
```

```python
spam=spam.where(pd.notnull(spam))
```

```python
spam.head()
```

```python
spam.shape
```

RENAMING

```python
spam.rename(columns={'v1':'Target'},inplace=True)
spam.rename(columns={'v2':'text'},inplace=True)
spam
```

DROPPING

```python
spam.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1)
```

```python
spam['Target'].value_counts()
```

```python
spam.loc[spam['Target']=='spam','Target']=0
spam.loc[spam['Target']=='ham','Target']=1
```

```python
x=spam['text']
y=spam['Target']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

TFIDF VECTORIZATION

```python
feature=TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)
```

```python
x_train_features=feature.fit_transform(x_train)
x_test_features=feature.transform(x_test)
```

```python
print(x_train_features)
```

```python
print(x_test_features)
```

```python
y_train=y_train.astype('int')
y_test=y_test.astype('int')
```

MODEL TRAINING

LOGISTIC REGRESSION

```python
model=LogisticRegression()
model.fit(x_train_features,y_train)
```

```python
pred1=model.predict(x_test_features)
accuracy_score(y_test,pred1)
```

```python
from sklearn import svm
svm=svm.SVC()
svm.fit(x_train_features,y_train)
```

```python
pred2=svm.predict(x_test_features)
accuracy_score(pred2,y_test)
```

RANDOM FOREST CLASSIFIER

```python
from sklearn.ensemble import RandomForestClassifier
model1=RandomForestClassifier()
model1.fit(x_train_features,y_train)
```

```python
RF=model1.predict(x_test_features)
accuracy_score(RF,y_test)
```

```python
print("LOGISTIC REGRESSION",accuracy_score(y_test,pred1))
print("SVM",accuracy_score(pred2,y_test))
print("RANDOM FOREST CLASSIFIER",accuracy_score(RF,y_test))
```
